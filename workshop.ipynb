```python
import pandas as pd
```

```python
df=pd.read_csv('/content/sample_data/E-commerce-data.csv')
```

```python
df
```

```python
df.head(10)
```

```python
df['Remarks'].unique()
```

```python
effective_df=df[df['Remarks']!='Verified Orders']
```

```python
effective_df
```

```python
total_count=effective_df['Count'].sum()
```

```python
total_count
```

```python
effective_df.groupby(['Remarks']).sum().plot(kind='pie',y='Count',autopct='%1.2f%%')
```

```python
df2=effective_df.groupby(['Period','Remarks']).agg({'Count':['sum']})
```

```python
df2
```

```python
import matplotlib.pyplot as plt
df2.reset_index(inplace=True)
plt.figure(figsize=(15,7))
for name, group in df2.groupby('Remarks'):
  plt.plot(group['Period'],group['Count']['sum'],label=name)

plt.title("Line Graph order remarks by period")
plt.xlabel("Period")
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
df3=effective_df.groupby(['Channel','Remarks']).agg({'Count':['sum']})
```

```python
df3
```

```python
df3.unstack()
```

```python
df3.unstack().plot(kind='bar')
plt.title("Bar Graph order remarks by CHANNEL")
plt.xlabel("Channel")
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
df4=effective_df.groupby(['State']).agg({'Count':['sum']})
```

```python
df4
```

```python
df4.plot(kind='barh')
plt.title("Barh Graph order remarks by state")
plt.xlabel("State")
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
df5=effective_df.groupby(['Dest Zone','Remarks']).agg({'Count':['sum']})
# print(df5)

df5.unstack().plot(kind='bar') #, stacke=true
plt.title("Bar Graph order remarks by Dest Zone")
plt.xlabel("Destination")
plt.ylabel('Count')
plt.legend()
plt.show()
```
